```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import requests
from bs4 import BeautifulSoup as bs
import os

# Reading the input file
df = pd.read_excel('/Users/suvankarmaity/Downloads/Meghla Internship Project/Input.xlsx')

# Starting webscraping
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

# Define the desired folder path
folder_path = "/Users/suvankarmaity/Downloads/Meghla Internship Project/extracted_data"

for index, row in df.iterrows():
    url = row['URL']
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = bs(response.content, "html.parser")
        H1 = soup.find_all('h1')[0].text
        # Convert heading into a string
        H1 = str(H1)
        # Content
        content = soup.find_all('div', class_='td-post-content tagdiv-type')[0]
        # Convert content into a string
        content = str(content)

        # Saving the content into a text file
        try:
            # Concatenate the strings
            combined_text = H1 + "  " + content

            # Select a specific value from the Series as the file name
            file_name = row['URL_ID'] + ".txt"

            # Join folder path with file name
            file_path = os.path.join(folder_path, file_name)

            # Save the combined content to the text file
            with open(file_path, "w") as file:
                file.write(combined_text)

            print(f"Combined text saved to '{file_path}' for URL_ID {row['URL_ID']}.")
        except Exception as e:
            print(f"Error saving content for URL_ID {row['URL_ID']}: {e}")
    else:
        print(f"Error accessing URL {url}. Status code: {response.status_code}")
```

```python
import os

def remove_stopwords(text, stopwords):
    # Tokenize the text into words
    words = text.split()
    # Remove stopwords
    cleaned_words = [word for word in words if word.lower() not in stopwords]
    # Join the remaining words back into a single string
    cleaned_text = ' '.join(cleaned_words)
    return cleaned_text

def read_stopwords(stopwords_file):
    with open(stopwords_file, 'r', encoding='utf-8') as file:
        stopwords = set([line.strip().lower() for line in file])
    return stopwords

def read_text_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

def clean_text_file(input_file, output_file, stopwords):
    # Read the content of the text file
    text = read_text_file(input_file)
    # Remove stopwords from the text
    cleaned_text = remove_stopwords(text, stopwords)
    # Optionally, write the cleaned text back to the file
    with open(output_file, 'w', encoding='utf-8') as file:
        file.write(cleaned_text)

# Define input and output folders
input_folder = "/Users/suvankarmaity/Downloads/Meghla Internship Project/extracted_data"
output_folder = "/Users/suvankarmaity/Downloads/Meghla Internship Project/removed_stopwords_data"
stopwords_file = "/Users/suvankarmaity/Downloads/Meghla Internship Project/stop_words.txt"

# Read the stopwords list
stopwords = read_stopwords(stopwords_file)

# Iterate over files in the input folder
for file_name in os.listdir(input_folder):
    if file_name.endswith('.txt'):
        input_file_path = os.path.join(input_folder, file_name)
        output_file_path = os.path.join(output_folder, file_name)
        # Clean the text file using stopwords
        clean_text_file(input_file_path, output_file_path, stopwords)
        print(f"Processed: {file_name}")
```

# Calculating Positive, Negative, Polarity, Subjectivity Score

```python


# Function to count the number of words in a file
def count_words(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()
        word_count = len(content.split())
    return word_count

# Function to load positive words from file
def load_words_from_file(file_path):
    with open(file_path, 'r') as f:
        return set(word.strip().lower() for word in f)

# Function to count positive or negative words in a file
def count_words_in_file(file_path, words):
    with open(file_path, 'r') as f:
        content = f.read().lower()
        return sum(1 if word in words else 0 for word in content.split())

# Calculate and print scores for each file
def calculate_and_print_scores(folder_path, positive_file, negative_file):
    positive_words = load_words_from_file(positive_file)
    negative_words = load_words_from_file(negative_file)

    for file_name in os.listdir(folder_path):
        if file_name.endswith('.txt'):
            file_path = os.path.join(folder_path, file_name)
            positive_score = count_words_in_file(file_path, positive_words)
            negative_score = count_words_in_file(file_path, negative_words)
            polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)
            total_words = count_words(file_path)
            subjectivity_score = (positive_score + negative_score) / (total_words + 0.000001)
            print(f"File: {file_name}, Positive Score: {positive_score}, Negative Score: {negative_score },Polarity Score: {polarity_score}, Subjectivity Score: {subjectivity_score:.6f}")

# Define paths and call the function
folder_path = "/Users/suvankarmaity/Downloads/Meghla Internship Project/removed_stopwords_data"
positive_file = "/Users/suvankarmaity/Downloads/Meghla Internship Project/positive_dict.txt"
negative_file = "/Users/suvankarmaity/Downloads/Meghla Internship Project/negative_dict.txt"

calculate_and_print_scores(folder_path, positive_file, negative_file)
```

# 2. Analysis of Readibility

```python


# Function to count the number of complex words in a list of words
def count_complex_words(words):
    cmu_dict = cmudict.dict()
    complex_word_count = 0
    for word in words:
        syllables = syllable_count(word, cmu_dict)
        if syllables > 2:
            complex_word_count += 1
    return complex_word_count

# Function to count the number of syllables in a word using the CMU Pronouncing Dictionary
def syllable_count(word, cmu_dict):
    if word.lower() in cmu_dict:
        return max([len(list(y for y in x if y[-1].isdigit())) for x in cmu_dict[word.lower()]])
    else:
        return 0

# Function to calculate the average sentence length
def calculate_average_sentence_length(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()
        sentences = sent_tokenize(content)
        words = word_tokenize(content)
        return len(words) / len(sentences) if len(sentences) > 0 else 0

# Function to calculate the percentage of complex words
def calculate_percentage_complex_words(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()
        words = word_tokenize(content)
        total_words = len(words)
        complex_word_count = count_complex_words(words)
        if total_words == 0:
            return 0  # Prevent division by zero
        else:
            return (complex_word_count / total_words) * 100

# Function to calculate the Fog Index for a file
def calculate_fog_index(file_path):
    average_sentence_length = calculate_average_sentence_length(file_path)
    percentage_complex_words = calculate_percentage_complex_words(file_path)
    fog_index = 0.4 * (average_sentence_length + percentage_complex_words)
    return fog_index

# Function to print Average Sentence Length, Percentage of Complex words, and Fog Index for each file
def analysis_of_readability(folder_path):
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if os.path.isfile(file_path):
            average_sentence_length = calculate_average_sentence_length(file_path)
            percentage_complex_words = calculate_percentage_complex_words(file_path)
            fog_index = calculate_fog_index(file_path)
            print(f"File: {file_name}, Average Sentence Length: {average_sentence_length:.2f}, Percentage of Complex Words: {percentage_complex_words:.2f}%, Fog Index: {fog_index:.2f}")

# Define the path to the folder containing the files
folder_path = "/Users/suvankarmaity/Downloads/Meghla Internship Project/removed_stopwords_data"

# Print Average Sentence Length, Percentage of Complex words, and Fog Index for each file in the folder
analysis_of_readability(folder_path)
```

# 3. Average Number of Words Per Sentence


```python
import os
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

# Function to count the number of words and sentences in a file
def count_words_and_sentences(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()
        sentences = sent_tokenize(content)
        words = word_tokenize(content)
        return len(words), len(sentences)

# Function to calculate the average number of words per sentence
def calculate_average_words_per_sentence(file_path):
    total_words, total_sentences = count_words_and_sentences(file_path)
    if total_sentences == 0:
        return 0  # Prevent division by zero
    else:
        return total_words / total_sentences

# Function to print Average Number of Words Per Sentence for each file
def print_average_words_per_sentence_for_files(folder_path):
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if os.path.isfile(file_path):
            average_words_per_sentence = calculate_average_words_per_sentence(file_path)
            print(f"File: {file_name}, Average Number of Words Per Sentence: {average_words_per_sentence:.2f}")

# Define the path to the folder containing the files
folder_path = "/Users/suvankarmaity/Downloads/Meghla Internship Project/removed_stopwords_data"

# Print Average Number of Words Per Sentence for each file in the folder
print_average_words_per_sentence_for_files(folder_path)
```

```python
import os
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize

# Function to count the number of words and sentences in all files in a folder
def count_words_and_sentences_in_folder(folder_path):
    total_words = 0
    total_sentences = 0
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if os.path.isfile(file_path):
            with open(file_path, "r", encoding="utf-8") as file:
                content = file.read()
                total_words += len(word_tokenize(content))
                total_sentences += len(sent_tokenize(content))
    return total_words, total_sentences

# Function to calculate the average number of words per sentence for the whole folder
def calculate_average_words_per_sentence_for_folder(folder_path):
    total_words, total_sentences = count_words_and_sentences_in_folder(folder_path)
    if total_sentences == 0:
        return 0  # Prevent division by zero
    else:
        return total_words / total_sentences

# Define the path to the folder containing the files
folder_path = "/Users/suvankarmaity/Downloads/Meghla Internship Project/removed_stopwords_data"

# Calculate the average number of words per sentence for the whole folder
average_words_per_sentence = calculate_average_words_per_sentence_for_folder(folder_path)
print(f"Average Number of Words Per Sentence for the folder: {average_words_per_sentence:.2f}")
```

# 4. count the number of complex words

```python
import os
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import cmudict

# Download the CMU Pronouncing Dictionary
nltk.download('cmudict')

# Function to count the number of complex words in a list of words
def count_complex_words(words):
    cmu_dict = cmudict.dict()
    complex_word_count = 0
    for word in words:
        syllables = syllable_count(word, cmu_dict)
        if syllables > 2:
            complex_word_count += 1
    return complex_word_count

# Function to count the number of syllables in a word using the CMU Pronouncing Dictionary
def syllable_count(word, cmu_dict):
    if word.lower() in cmu_dict:
        return max([len(list(y for y in x if y[-1].isdigit())) for x in cmu_dict[word.lower()]])
    else:
        return 0

# Function to calculate the count of complex words for each file
def calculate_complex_word_count_for_files(folder_path):
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if os.path.isfile(file_path):
            with open(file_path, "r", encoding="utf-8") as file:
                content = file.read()
                words = word_tokenize(content)
                complex_word_count = count_complex_words(words)
                print(f"File: {file_name}, Complex Word Count: {complex_word_count}")

# Define the path to the folder containing the files
folder_path = "/Users/suvankarmaity/Downloads/Meghla Internship Project/removed_stopwords_data"

# Calculate the count of complex words for each file in the folder
calculate_complex_word_count_for_files(folder_path)
```

# 5. Cleaned Word Count


```python
import os
import nltk
from nltk.tokenize import word_tokenize
from string import punctuation
from nltk.corpus import stopwords

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')

# Function to count the total cleaned words in a file
def count_cleaned_words(file_path):
    stop_words = set(stopwords.words('english'))
    cleaned_word_count = 0

    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()
        words = word_tokenize(content)
        
        # Remove punctuations and stop words
        cleaned_words = [word.lower() for word in words if word.lower() not in stop_words and word.lower() not in punctuation]
        
        cleaned_word_count = len(cleaned_words)

    return cleaned_word_count

# Function to calculate the total cleaned words for each file
def calculate_total_cleaned_words_for_files(folder_path):
    total_cleaned_word_count = 0
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if os.path.isfile(file_path):
            cleaned_word_count = count_cleaned_words(file_path)
            total_cleaned_word_count += cleaned_word_count
            print(f"File: {file_name}, Cleaned Word Count: {cleaned_word_count}")

    print(f"Total Cleaned Word Count for all files: {total_cleaned_word_count}")

# Define the path to the "removed_stopwords_data" folder
folder_path = "/Users/suvankarmaity/Downloads/Meghla Internship Project/removed_stopwords_data"

# Calculate the total cleaned words for each file in the folder
calculate_total_cleaned_words_for_files(folder_path)
```

# 6. Syllable Count Per Word


```python
import os
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import cmudict

# Download the CMU Pronouncing Dictionary
nltk.download('cmudict')

# Function to count the number of syllables in each word
def count_syllables(word):
    cmu_dict = cmudict.dict()
    # Check if the word is in the CMU Pronouncing Dictionary
    if word.lower() in cmu_dict:
        syllable_count = 0
        # Iterate through each phoneme in the pronunciation of the word
        for phoneme in cmu_dict[word.lower()][0]:
            # Count the number of digits (indicating stress) in the phoneme
            if any(char.isdigit() for char in phoneme):
                syllable_count += 1
        return syllable_count
    else:
        return 0  # Return 0 if word not found in CMU Pronouncing Dictionary

# Function to count the total syllable count per file
def count_total_syllables_in_text(file_path):
    total_syllable_count = 0
    with open(file_path, "r", encoding="utf-8") as file:
        content = file.read()
        words = word_tokenize(content)
        for word in words:
            # Handle exceptions for words ending with "es" or "ed"
            if not (word.endswith("es") or word.endswith("ed")):
                total_syllable_count += count_syllables(word)
    return total_syllable_count

# Function to count the total syllable count per file for all files in the folder
def count_total_syllables_in_files(folder_path):
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if os.path.isfile(file_path):
            total_syllable_count = count_total_syllables_in_text(file_path)
            print(f"File: {file_name}, Total Syllable Count: {total_syllable_count}")

# Define the path to the folder containing the files
folder_path = "/Users/suvankarmaity/Downloads/Meghla Internship Project/removed_stopwords_data"

# Count the total syllable count per file for all files in the folder
count_total_syllables_in_files(folder_path)
```

# 7. Personal Pronouns


```python
import os
import re

# Function to count personal pronouns in text
def count_personal_pronouns(text):
    personal_pronouns = ["I", "we", "my", "ours", "us"]
    # Regular expression pattern to match personal pronouns
    pattern = r'\b(?:' + '|'.join(personal_pronouns) + r')\b'
    # Find all matches of personal pronouns in the text
    matches = re.findall(pattern, text, flags=re.IGNORECASE)
    return len(matches)

# Function to count personal pronouns in each file
def count_personal_pronouns_in_files(folder_path):
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if os.path.isfile(file_path):
            with open(file_path, "r", encoding="utf-8") as file:
                content = file.read()
                pronoun_count = count_personal_pronouns(content)
                print(f"File: {file_name}, Personal Pronoun Count: {pronoun_count}")

# Define the path to the folder containing the files
folder_path = "/Users/suvankarmaity/Downloads/Meghla Internship Project/removed_stopwords_data"

# Count personal pronouns in each file in the folder
count_personal_pronouns_in_files(folder_path)
```

# 8. Average Word Length


```python
import os
import nltk
from nltk.tokenize import word_tokenize

# Function to calculate average word length in text
def calculate_average_word_length(text):
    words = word_tokenize(text)
    total_characters = sum(len(word) for word in words)
    total_words = len(words)
    if total_words == 0:
        return 0  # Prevent division by zero
    else:
        return total_characters / total_words

# Function to calculate average word length for each file
def calculate_average_word_length_in_files(folder_path):
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if os.path.isfile(file_path):
            with open(file_path, "r", encoding="utf-8") as file:
                content = file.read()
                average_word_length = calculate_average_word_length(content)
                print(f"File: {file_name}, Average Word Length: {average_word_length:.2f}")

# Define the path to the folder containing the files
folder_path = "/Users/suvankarmaity/Downloads/Meghla Internship Project/removed_stopwords_data"

# Calculate average word length for each file in the folder
calculate_average_word_length_in_files(folder_path)
```
